# Creating, Converting, and Plotting

Below I find the mean of all dates, store them in a list, then store all dates inside 'date', create a new dataframe called overallsteps and add two columns: mean and date. This helps to narrow the data to relevant information I can later plot. Afterwards I check the data types which helps me understand what information I'm working with and what I need to change to make a successful time series plot. I convert the 'date' to 'datetime' so that I will be able to make a plot. Afterwards I use '.info' to check that the conversion was successful before creating my mean steps by date plot. 

```python
col = df1.columns
overAllSteps = pd.DataFrame()
l = []
date = []
for i in col:
    l.append( df1[i].mean())
    date.append(i)

overAllSteps['mean'] = l[1:]
overAllSteps['date'] = date[1:]
overAllSteps.head()
```

```python
overAllSteps['date']= pd.to_datetime(overAllSteps['date'])  
```

```python
overAllSteps.info() 
```

```python
#creating a time series plot
plt.rcParams["figure.figsize"] = [25,10] 
overAllSteps.plot(x='date', y='mean')
plt.title("Mean steps per date over all participants") #set the title
plt.xlabel("Date") #set the x-label
plt.ylabel("Mean steps") #set the y-label
```

# Analysis of Quantitative Variable 

For the quantitative variable analysis I chose to focus on weight for my two distribution graphs. 

```python
sns.scatterplot(data=df, x="weight", y = "id")
plt.title("Weight distribution of participants")
plt.xlabel("Weight")
plt.ylabel("ID")
plt.show()
```

```python
sns.distplot(df["weight"], hist=False, kde=True) 
plt.title("Weight distribution of participants")
plt.xlabel("Weight")
plt.ylabel("Proportion")
plt.show()
```

## Summary statistics & Verbal description

```python
# calculating weight statistics 
print('median: ' + str(float(df[['weight']].median())))
print('mode: ' + str(df[['weight']].mode()['weight'][0]))
print('mean: ' + str(float(df[['weight']].mean())))
print('standard deviation: ' + str(float(df[['weight']].std())))
print('variance: ' + str(float(df[['weight']].var())))
```

The center of the weight distribution is around 70. The standard deviation is about 26. The majority of weights are  over 50 and under 100 kg. There are some variables that are naturally higher and two outliers that potentially represent errors in the data with weights listed at 700 kg and 0 kg.
